```python
import numpy as np
import pandas as pd
path= '/content/antropometria-dataset-2.csv'
```

```python
df= pd.read_csv(path)
```

```python
df
```

```python
pivottable= pd.pivot_table(df, values= 'peso', index= 'desc_ent', aggfunc=min, columns='sexo')
pivottable
```

Average weight of people from SINALOA

```python
pivottable= pd.pivot_table(df, values= 'peso', index= 'desc_ent', aggfunc='mean')
pivottable
```

Median weight(peso) of sex2 of  YUCATAN es

```python
pivottable= pd.pivot_table(df, values= 'peso', index= 'desc_ent', aggfunc='median', columns='sexo')
pivottable
```
